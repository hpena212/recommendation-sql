```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
data = pd.read_csv(r"C:\Users\hecto\Documents\Z123\rawcleaned_customer_data.csv")

# Plot churn distribution
sns.countplot(x='Churn', data=data)
plt.title('Churn Distribution')
plt.show()

```

```python
# Convert 'Churn' to numeric (0 for No, 1 for Yes)
data['Churn'] = data['Churn'].map({'No': 0, 'Yes': 1})

# Calculate churn rate for each contract type
churn_by_contract = data.groupby('Contract')['Churn'].mean()

# Plot bar chart
plt.bar(churn_by_contract.index, churn_by_contract.values, color=['blue', 'orange', 'green'])
plt.xlabel('Contract Type')
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Contract Type')
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=data)
plt.title('Churn by Contract')
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=data)
plt.title('Monthly Charges by Churn')
plt.show()
```

```python
sns.countplot(x='InternetService', hue='Churn', data=data)
plt.title('Churn by Internet Service')
plt.show()
```

```python
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=data)
plt.title('Tenure vs Monthly Charges with Churn')
plt.show()
```
